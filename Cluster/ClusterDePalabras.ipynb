
## Cluster de palabras

### 1. Usando Word2vec

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from sklearn.manifold import TSNE

filename = 'GoogleNews-vectors-negative300.bin'
model = KeyedVectors.load_word2vec_format(filename, binary=True)
tsne = TSNE(n_components=2)

def dataImport(filename):
    info = []
    data = open(filename)
    for r in data:
        r = r.replace('\n', '')
        info.append(r)
    return info


sports = dataImport('sports_words.txt')
food = dataImport('food_words.txt')
weather = dataImport('weather_words.txt')
print(model)
```

```python
import matplotlib.pyplot as plt

def vector(data):
    vector = []
    for xs in data:
        try:
            vector.append(model[xs])
        except:
            #print(xs)
            pass
    vector_em = tsne.fit_transform(vector)
    return vector_em

vector_sport = vector(sports)
vector_food = vector(food)
vector_weather = vector(weather)

plt.figure(figsize=(22.2, 11.4))
plt.scatter(vector_sport[:,0],vector_sport[:,1],color="green")
plt.scatter(vector_food[:,0],vector_food[:,1], color="red")
plt.scatter(vector_weather[:,0],vector_weather[:,1], color="blue")
plt.show
```
